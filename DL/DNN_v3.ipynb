```python
import pandas as pd
import numpy as np

# データの読み込み
df = pd.read_csv('../data/BTC_1m.csv', index_col=0)
df
```

```python
# 対数収益率
df['return'] = np.log(df['Close'] / df['Close'].shift(1))

# 対数収益率が正の場合（価格が上昇した場合）には１を，そうでない場合は０を割り当てる
df['direction'] = np.where(df['return'] > 0, 1, 0)
df
```

```python
lags = 5

cols = []
for lag in range(1, lags + 1):
    col = f'lag_{lag}'
    df[col] = df['return'].shift(lag)
    cols.append(col)
df.dropna(inplace=True)
df.round(4).tail()
```

```python
# データをtrainとtestに分ける
# データフレームの行数を計算
n_rows = df.shape[0]

# 訓練データとテストデータに分割する行のインデックスを計算
split_index = int(n_rows * 0.8)

# 訓練データとテストデータに分割
train_df = df[:split_index]
test_df = df[split_index:]

# 結果を表示
print(f"全データの数: {n_rows}")
print(f"訓練データの数: {train_df.shape[0]}")
print(f"テストデータの数: {test_df.shape[0]}")
train_df
```

```python
# ガウス標準化(Zスコア正規化)によって特徴量データを標準化
mu, std = train_df.mean(), train_df.std()
train_df_ = (train_df - mu) / std
test_df_ = (test_df - mu) / std
# train_df_.columns
```

```python
%%time
# Kerasを使ってDNN
# ハイパーパラメータ調整追加
import numpy as np
import tensorflow as tf
from sklearn.model_selection import GridSearchCV
from scikeras.wrappers import KerasClassifier
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam, RMSprop
import random

optimizer = Adam(learning_rate=0.0001) # Adam最適化アルゴリズムを使用し，学習率を0.001に設定

def set_seeds(seed=100): # 乱数生成のシードを設定，再現性の確保
    random.seed(seed)
    np.random.seed(seed)
    tf.random.set_seed(100)

set_seeds()

input_dim = 28

# Kerasのモデルを作成する関数
def create_model(optimizer='adam', activation='relu'):
    model = Sequential()
    model.add(Dense(units=12, input_shape=(input_dim,), activation=activation))
    model.add(Dense(units=8, activation=activation))
    model.add(Dense(units=1, activation='sigmoid'))
    model.compile(optimizer=optimizer, loss='binary_crossentropy', metrics=['accuracy'])
    return model

# モデルとグリッドサーチの設定
model = KerasClassifier(model=create_model, verbose=0)

# ハイパーパラメータの探索範囲
param_grid = {
    'model__optimizer': ['adam', 'rmsprop'],  # 'model__' プレフィックスを追加
    'model__activation': ['relu', 'tanh'],
    'batch_size': [16, 32, 64],
    'epochs': [10, 20]
}

grid = GridSearchCV(estimator=model, param_grid=param_grid, n_jobs=-1, cv=3)

# グリッドサーチの実行
grid_result = grid.fit(train_df[['Open', 'High', 'Low', 'Close', 'Volume', 'Dividends', 'Stock Splits','AD', 'ma15', 'ma30', 'ma75', 
          'gc', 'dc', 'upper2', 'lower2', 'macd','macd_signal', 'hist', 'rsi14', 'rsi28', 'slowK', 'slowD', 'return',
          'lag_1', 'lag_2', 'lag_3', 'lag_4', 'lag_5']], train_df['direction'])

# 結果の表示
print("Best: %f using %s" % (grid_result.best_score_, grid_result.best_params_))
means = grid_result.cv_results_['mean_test_score']
stds = grid_result.cv_results_['std_test_score']
params = grid_result.cv_results_['params']
for mean, stdev, param in zip(means, stds, params):
    print("%f (%f) with: %r" % (mean, stdev, param))

# model = Sequential()
# model.add(Dense(64, activation='relu', input_shape=(28,))) # 64個のニューロン，reluを活性化関数としている
# model.add(Dense(64, activation='relu'))
# model.add(Dense(1, activation='sigmoid')) # 出力層で，1個のニューロンを持つ sigmoid活性化関数を使用し，二項分類問題に対応
# model.compile(optimizer=optimizer, loss='binary_crossentropy', metrics=['accuracy'])
```
