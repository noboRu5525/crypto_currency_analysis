```python
import pandas as pd
import numpy as np

# データの読み込み
df = pd.read_csv('../data/BTC_1m.csv', index_col=0)
df
```

```python
# 対数収益率
df['return'] = np.log(df['Close'] / df['Close'].shift(1))

# 対数収益率が正の場合（価格が上昇した場合）には１を，そうでない場合は０を割り当てる
df['direction'] = np.where(df['return'] > 0, 1, 0)
df
```

```python
lags = 5

cols = []
for lag in range(1, lags + 1):
    col = f'lag_{lag}'
    df[col] = df['return'].shift(lag)
    cols.append(col)
df.dropna(inplace=True)
df.round(4).tail()
```

```python
# Kerasを使ってDNN
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam, RMSprop
import random

optimizer = Adam(learning_rate=0.0001) # Adam最適化アルゴリズムを使用し，学習率を0.001に設定

def set_seeds(seed=100): # 乱数生成のシードを設定，再現性の確保
    random.seed(seed)
    np.random.seed(seed)
    tf.random.set_seed(100)

set_seeds()

model = Sequential()
model.add(Dense(64, activation='relu', input_shape=(28,))) # 64個のニューロン，reluを活性化関数としている
model.add(Dense(64, activation='relu'))
model.add(Dense(1, activation='sigmoid')) # 出力層で，1個のニューロンを持つ sigmoid活性化関数を使用し，二項分類問題に対応
model.compile(optimizer=optimizer, loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# データをtrainとtestに分ける
# データフレームの行数を計算
n_rows = df.shape[0]

# 訓練データとテストデータに分割する行のインデックスを計算
split_index = int(n_rows * 0.8)

# 訓練データとテストデータに分割
train_df = df[:split_index]
test_df = df[split_index:]

# 結果を表示
print(f"全データの数: {n_rows}")
print(f"訓練データの数: {train_df.shape[0]}")
print(f"テストデータの数: {test_df.shape[0]}")
train_df
```

```python
# ガウス標準化(Zスコア正規化)によって特徴量データを標準化
mu, std = train_df.mean(), train_df.std()
train_df_ = (train_df - mu) / std
test_df_ = (test_df - mu) / std
train_df_.columns
```

```python
%%time
#モデルの訓練
model.fit(train_df[['Open', 'High', 'Low', 'Close', 'Volume', 'Dividends', 'Stock Splits','AD', 'ma15', 'ma30', 'ma75', 
          'gc', 'dc', 'upper2', 'lower2', 'macd','macd_signal', 'hist', 'rsi14', 'rsi28', 'slowK', 'slowD', 'return',
          'lag_1', 'lag_2', 'lag_3', 'lag_4', 'lag_5']], 
          train_df['direction'], 
          epochs=50, 
          verbose=False, 
          validation_split=0.2, 
          shuffle=False) 
```

```python
res = pd.DataFrame(model.history.history) # 訓練の履歴情報を取得しデータフレームに変換，損失値（Loss），精度（Accuracy），検証損失値（Validation Loss），検証精度（Validation Accuracy）が含まれる
```

```python
res[['accuracy', 'val_accuracy']].plot(figsize=(10, 6), style='--') # 訓練ステップごとの訓練データと検証データに対するDNN分類機の精度
```

```python
model.evaluate(train_df_[['Open', 'High', 'Low', 'Close', 'Volume', 'Dividends', 'Stock Splits','AD', 'ma15', 'ma30', 'ma75', 
          'gc', 'dc', 'upper2', 'lower2', 'macd','macd_signal', 'hist', 'rsi14', 'rsi28', 'slowK', 'slowD', 'return',
          'lag_1', 'lag_2', 'lag_3', 'lag_4', 'lag_5']], train_df['direction'])
```

```python
# イン・サンプルでマーケットの方向性を予測
pred = np.where(model.predict(train_df_[['Open', 'High', 'Low', 'Close', 'Volume', 'Dividends', 'Stock Splits','AD', 'ma15', 'ma30', 'ma75', 
          'gc', 'dc', 'upper2', 'lower2', 'macd','macd_signal', 'hist', 'rsi14', 'rsi28', 'slowK', 'slowD', 'return',
          'lag_1', 'lag_2', 'lag_3', 'lag_4', 'lag_5']]) > 0.5, 1, 0)
pred[:30].flatten()
```

```python
# 予測値をロングショートのポジション「+1」，「-1」に変換
train_df['prediction'] = np.where(pred > 0, 1, -1)
```

```python
# ポジションを与えて戦略のリターンを計算
train_df['strategy'] = (train_df['prediction'] * train_df['return'])
train_df[['return', 'strategy']].sum().apply(np.exp)
```

```python
# 戦略のパファーマンスをプロットし，ベンチマークのパフォーマンス（イン・サンプル）と比較する
train_df[['return', 'strategy']].cumsum().apply(np.exp).plot(figsize=(10, 6))
```

```python
model.evaluate(test_df_[['Open', 'High', 'Low', 'Close', 'Volume', 'Dividends', 'Stock Splits','AD', 'ma15', 'ma30', 'ma75', 
          'gc', 'dc', 'upper2', 'lower2', 'macd','macd_signal', 'hist', 'rsi14', 'rsi28', 'slowK', 'slowD', 'return',
          'lag_1', 'lag_2', 'lag_3', 'lag_4', 'lag_5']], test_df['direction'])
```

```python
pred = np.where(model.predict(test_df_[['Open', 'High', 'Low', 'Close', 'Volume', 'Dividends', 'Stock Splits','AD', 'ma15', 'ma30', 'ma75', 
          'gc', 'dc', 'upper2', 'lower2', 'macd','macd_signal', 'hist', 'rsi14', 'rsi28', 'slowK', 'slowD', 'return',
          'lag_1', 'lag_2', 'lag_3', 'lag_4', 'lag_5']]) > 0.5, 1, 0)
```

```python
test_df['prediction'] = np.where(pred > 0, 1, -1)
```

```python
test_df['prediction'].value_counts()
```

```python
test_df['strategy'] = (test_df['prediction'] * test_df['return'])
test_df[['return', 'strategy']].sum().apply(np.exp)
```

```python
# 戦略のパファーマンスをプロットし，ベンチマークのパフォーマンス（アウト・サンプル）と比較する
test_df[['return', 'strategy']].cumsum().apply(np.exp).plot(figsize=(10, 6))
```
